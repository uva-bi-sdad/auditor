```python
def transform_geoid(raw):
    if raw in d:
        return d[raw]['hd_geoid']
    else:
        result = process.extract(raw, d.keys(), limit=1)[0]
#         print(result)
#         print(result[0] in d)
        if result[1] >= 90:
            return d[result[0]]['hd_geoid']
        else:
            return raw
```
